```python
import pandas as pd
dataframe = pd.DataFrame()
```

```python
from google.colab import files
uploaded = files.upload()
df = pd.read_csv('iris.csv')

# Show first two rows
print(df.head(2))

# Show last two rows
print(df.tail(2))

# Show dimensions
print("Dimensions: {}".format(df.shape))

# Show statistics
print(df.describe())
```

```python
# Select the first row
print(df.iloc[0])

# Select three rows
print(df.iloc[:3])

# All rows up to and including the fourth row
print(df.iloc[:4])
```

```python
# Example: select top two rows where alcohol > 10 (change the column name to your dataset)
df_selected = df[df['sepal_length'] > 1].head(2)
print(df_selected)

# Multiple conditions (e.g., alcohol > 10 and quality > 5)
df_multiple = df[(df['sepal_length'] > 1) & (df['sepal_width'] > 5)].head(2)
print(df_multiple)
```

```python
# Replace specific value
df['sepal_length'].replace(3, 1, inplace=True)
print(df.head(2))

# Replace multiple values
df['sepal_length'].replace({3: 1, 8: 2}, inplace=True)
print(df.head(2))

# Replace in entire dataframe
df.replace(1, 'One', inplace=True)
print(df.head(2))
```

```python
# Rename one column
df.rename(columns={'sepal_length': 'sepallen'}, inplace=True)
print(df.head(2))

# Rename multiple columns
df.rename(columns={'sepal_width': 'sepalwid'}, inplace=True)
print(df.head(2))
```

```python
print("Maximum: {}".format(df['sepallen'].max()))
print("Minimum: {}".format(df['sepallen'].min()))
print("Mean: {}".format(df['sepallen'].mean()))
print("Sum: {}".format(df['sepallen'].sum()))
print("Count: {}".format(df['sepallen'].count()))

# Whole dataframe
print("Variance:\n{}".format(df.var(numeric_only=True)))
print("Standard Deviation:\n{}".format(df.std(numeric_only=True)))
print("Kurtosis:\n{}".format(df.kurt(numeric_only=True)))
print("Skewness:\n{}".format(df.skew(numeric_only=True)))
```

```python
print(df.groupby('sepalwid').count())
```

```python
print(df['sepalwid'].unique())
print(df['sepalwid'].value_counts())
```

```python
# Select missing values, show 2 rows
print(df[df.isnull().any(axis=1)].head(2))
```

```python
df_dropped = df.drop('sepallen', axis=1)
print(df_dropped.head(2))
```

```python
df_filtered = df[df['sepalwid'] != 6]
print(df_filtered.head(2))

# Delete row by index
df_no_first = df.drop(df.index[0])
print(df_no_first.head(2))
```

```python
print(df.groupby('sepalwid').mean(numeric_only=True))
print(df.groupby('sepalwid')['petal_width'].count())
```

```python
for val in df['sepalwid'][:2]:
    print(val)

# List comprehension
vals_upper = [str(val).upper() for val in df['sepalwid'][:2]]
print(vals_upper)
```

```python
def stringify(x):
    return str(x)

df['petal_width'] = df['sepalwid'].apply(stringify)
print(df[['sepalwid', 'petal_width']].head(2))
```
